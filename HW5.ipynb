```python
#Question 8

import numpy as np
from scipy import stats
import plotly.express as px
import plotly.graph_objects as go

# Set the parameters
n_per_sample = 80  # Number of students per sample
n_sample_simulations = 10000  # Number of samples to simulate
simulated_proportion = np.zeros(n_sample_simulations)
H0_p = 0.5
np.random.seed(42)

# Simulate the samples
for i in range(n_sample_simulations):
    # Simulate the tastes
    n_correct = np.random.choice(['T', 'F'], p=[H0_p, 1 - H0_p], size=n_per_sample)
    simulated_proportion[i] = (n_correct == 'T').mean()

# Create the histogram using plotly.express
fig = px.histogram(simulated_proportion, nbins=30,
                   title='Sampling Distribution of Proportion of Correct Tastes',
                   labels={'value': 'Proportion of Corrects', 'count': 'Frequency'})
fig.update_traces(name='Simulated Proportions')
fig.show(render='png')

# Observed proportion statistic
observed_proportion_statistic = 49 / 80
print("If we observed a proportion of", observed_proportion_statistic, "correct out of", n_per_sample, "students")

# One-tailed p-value calculation
# Calculate the p-value: proportion of simulations greater than observed proportion
p_value_one_tailed = (simulated_proportion >= observed_proportion_statistic).mean()

print("The one-tailed p-value is:", p_value_one_tailed)
print("This represents the probability of observing a proportion as extreme or more extreme than", observed_proportion_statistic, "if H0 is true.")

# Plot the histogram again with the observed proportion
fig = px.histogram(simulated_proportion, nbins=30,
                   title='Sampling Distribution of Proportion of Correct Tastes',
                   labels={'value': 'Proportion of Corrects', 'count': 'Frequency'})
fig.update_traces(name='Simulated Proportions<br>under the assumption<br>that H0 is true<br>')

# Add vertical line for observed proportion
fig.add_trace(go.Scatter(x=[observed_proportion_statistic] * 2,
                         y=[0, max(np.histogram(simulated_proportion, bins=30)[0])],
                         mode='lines', line=dict(color='red', width=2), name='Observed Proportion'))

# Mark the p-value area (right tail)
x_as_or_more_extreme = simulated_proportion[simulated_proportion >= observed_proportion_statistic]
fig.add_trace(go.Scatter(x=x_as_or_more_extreme, y=0 * x_as_or_more_extreme,
    mode='markers', marker=dict(color='red', size=10, symbol='x'),
    name='"as or more extreme" (one-tailed)'))

fig.show(render='png')


#Conclusion: as the p-value for the test is 0.0277<0.05, there is a very strong evidence that we reject the null hypothesis, 
# so we could say people can taste the order of the milk and tea pouring in the drink.
```
